```python
def list_operations():
    list = []
    size = int(input("enter size"))
    for i in range(size):
        val = int(input("enter the val"))
        list.append(val)

    while True:
        print("\nList Operations:")
        print("1. Insert an element")
        print("2. Delete an element")
        print("3. Find an element")
        print("4. Display list")
        print("5. Exit")

        choice = int(input("Enter your choice: "))

        if choice == 1:
            element = input("Enter element to insert: ")
            list.append(element)
            print(f"Element '{element}' inserted.")

        elif choice == 2:
            element = input("Enter element to delete: ")
            if element in list:
                list.remove(element)
                print(f"Element '{element}' deleted.")
            else:
                print("Element not found.")

        elif choice == 3:
            element = input("Enter element to find: ")
            if element in list:
                print(f"Element '{element}' found in the list.")
            else:
                print("Element not found.")

        elif choice == 4:
            print("Current List:", list)

        elif choice == 5:
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please try again.")
            
list_operations()
```

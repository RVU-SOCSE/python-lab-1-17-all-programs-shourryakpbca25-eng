```python
#switch_case
day=int(input("enter a day from 1 to 3"))
match day:  
    case 1:
        print("monday")
    case 2:
        print("tuesday")
    case 3:
        print("wednesday")
    case _:
        print("invalid input")
    
```

```python
ch = input("Enter a character: ")

match ch:
    case "a" | "e" | "i" | "o" | "u":
        print("Vowel")
    case _:
        print("Consonant")
```

```python
#lab 2
#factorial using recursion

def fact(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * fact(n - 1)
num= int(input("Enter a number: "))
if num < 0:
    print("Factorial is not possible for this number ")
else:
    res=factorial(num)
    print("The factorial of ", num,  "is ", res)
```

```python
#fibonacci sequence program
def fib(n):
    if n<=1:
        return n
    else:
        return fib(n-1)+fib(n-2)
print(fib(10))
```

```python
#word count in a sentence 
def lengthOfString(String):
    length = 0;
 
    for char in String:
        length += 1
    return length

abc=input("enter the string: ")

print(lengthOfString(abc))
```

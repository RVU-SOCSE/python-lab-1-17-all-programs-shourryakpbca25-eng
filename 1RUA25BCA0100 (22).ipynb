```python
import numpy as np

arr1=np.array([1,2,3,4,5,6])
arr2=np.array([3,6,9,2,4,8])

sum=arr1+arr2
product= arr1 * arr2

print(f"the sum of {arr1} and {arr2} is : ",sum)
print(f"the product of {arr1} and {arr2} is :",product)
```

```python
import numpy as np

arr1=np.array([[1,2,3],
              [4,6,3]])
arr2=np.array([[3,6,9],
              [2,4,8]])

sum=arr1+arr2
funcsum=np.add(arr1, arr2)

product=arr1*arr2
funcpro=np.multiply(arr1, arr2)

print(f"the sum of {arr1} and {arr2} is :", sum)
print(f"the sum of {arr1} and {arr2} is using function :", funcsum)

print(f"the product of {arr1} and {arr2} is :",product)
print(f"the product of {arr1} and {arr2} is :",funcpro)
```

```python
import numpy as np

arr1=np.array([[1,2,3],
              [4,6,3]])
arr2=np.array([[3,6,9],
              [2,4,8]])

sub=arr1-arr2
sub1=np.subtract(arr1, arr2)

print(f"the difference of {arr1} and {arr2} is : ", sub)
print(f"the difference using sum is: ", sub1)
```

```python
import numpy as np

mat1=np.array([[1,2,3],
              [4,6,3]])
mat2=np.array([[3,6],
              [2,4],
              [6,9]])

print(np.shape(mat1))
print(np.shape(mat2))

dot=np.dot(mat1, mat2)
trans=np.transpose(mat1)

print("the product is:",dot)
print(trans)
```

```python
import numpy as np

arr1=np.array([1,2,3,4,5,6])
arr2=np.array([3,6,9,2,4,8])

quo=np.divide(arr1, arr2)

print(f"The quotent of {arr1} and {arr2} is :", quo)
```

```python
import numpy as np

arr1=np.array([1,2,3,4,5,6])
arr2=np.array([3,6,9,2,4,8])

mo=np.mod(arr1, arr2)
pow=np.power(arr1, arr2)
abs=np.absolute(arr1, arr2)
neg=np.negative(arr1, arr2)

print("the modulus is:", mo)
print("The power is:", pow)
print("the absolute is ", abs)
print("the negative is:", neg)
```

```python
#LAB1
#simple calculator in python
#first program 

a=int(input("enter first number: "))
b=int(input("Enter the second number: "))

op=int(input("enter the operation to be performed: \n 1:add \n 2:subtract \n 3:divide \n 4:multiply \n "))

if (op==1):
    print(a, "+", b, "=", a+b)
elif(op==2):
    print(a, "-", b, "=", a-b )
elif(op==3):
    print(a, "/", b, "=", a/b )
elif(op==4):
    print(a, "*", b, "=", a*b )
else:
    print("not a given operation \n please choose an option from the given operations ")
```

```python
#first program using switch case

a=int(input("enter first number: "))
b=int(input("Enter the second number: "))
op=int(input("enter the operation to be performed: \n 1:add \n 2:subtract \n 3:divide \n 4:multiply \n "))

match op:
    case 1:
        print(a, "+", b, "=", a+b)
    case 2:
        print(a, "-", b, "=", a-b )
    case 3:
        print(a, "/", b, "=", a/b )
    case 4:
        print(a, "*", b, "=", a*b )
    case _:
         print("not a given operation \n please choose an option from the given operations ")
```

```python
#second program second try

num=int(input("enter the desired temperature: \n 1:celsius \n 2:farenheit \n "))

match num:
    case 1:
        far=int(input("enter the temperature in farenheit:"))
        cel=(far-32)/.18
        print("  The given temperature in celsius is :", cel)

    case 2:
        cel=int(input("Enter the temperature in celcius : "))
        far=(cel*1.8)+32
        print("  The given temperature in farenheit is :  ", far)

    case _:
        print("no conversion possible")
```

```python
#find the largest of three numbers 
#third program

a=int(input("Enter first number :"))
b=int(input("Enter second number :"))
c=int(input("Enter third number :"))

if (a>b and a>c):
    print(a, "is greater than", b, c)
elif(b>c and b>a):
    print(b, "is greater than ", a,c)
else:
    print(c, "is greatest")

if (a<b and a<c):
    print(a, "is smaller than", b, c)
elif(b<c and b<a):
    print(b, "is smaller than ", a,c)
else:
    print(c, "is smallest")
```

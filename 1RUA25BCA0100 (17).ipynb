```python
#to calculate the are and perimeter of a rectangle and square
n=int(input("Enter the choice \n 1:rectangle \n 2:square \n"))

if(n==1):
    print("Enter the lenght and breadth \n")
    len=int(input("Enter the lenght of the rectangle: \n"))
    bre=int(input("Enter the breadth of the rectangle: \n"))
    area=len*bre
    peri=2*len*bre
    print("area is",area)
    print("perimeter is",peri)
    
elif(n==2):
    print("Enter the values of the square")
    side=int(input("Enter the side of the square"))
    areas=side*side
    peris=4*side
    print("area is",areas)
    print("perimeter is",peris)
else:
    print("invalid input GO AWAY")
```
